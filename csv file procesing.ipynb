```python
import pandas as pd
```

```python
persons = pd.read_csv("illinois_doc_dataset/csv/person.csv", delimiter = ';')
```

```python
persons.head()
```

```python
persons.columns
```

```python
persons_proc = pd.DataFrame()
```

```python
persons_proc['id'] = persons['id']
persons_proc['weight'] = persons['weight']
persons_proc['height'] = persons['height']
persons_proc['sex'] = persons['sex']
persons_proc.head()
```

```python
persons_proc['BMI'] = (persons_proc['weight']/(persons_proc['height']**2))*703
```

```python
persons_proc.head()
```

```python
persons_proc.to_csv('persons_processed.csv', index = False)
```
