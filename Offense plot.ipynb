```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sentencing.csv', sep =';')
print(df.head())
```

```python
print("metadata:")
print(f"column names: {df.columns.tolist()}")
print(f"data types of each column: \n{df.dtypes}")
print(f"the first few rows of the dataset: \n{df.head()}")
```

```python
offense_column = 'offense'
offense_counts = df[offense_column].value_counts()

# Step 5: Limit to top N offenses (e.g., top 20 most common offenses)
top_n = 20
offense_counts_top = offense_counts.head(top_n)

# Step 6: Plot the distribution of offenses
plt.figure(figsize=(12, 8))  # Adjust the size of the plot
offense_counts_top.plot(kind='bar', color='lightcoral')
plt.title('Top 20 Offenses Committed by Inmates')
plt.xlabel('Offense')
plt.ylabel('Number of Inmates')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```
